```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("CALLS.csv")
Strikes = np.unique(df.Strike.values)
Maturities = np.unique(df.Maturity.values)
df = df.set_index(["Strike", "Maturity"]).unstack()
mat = df.values

T, K = np.meshgrid(Maturities, Strikes)
```

```python
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')

ax.plot_surface(K, T, mat,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
plt.show()
```
